```python
import os
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn.neural_network import MLPRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
import joblib

# Ensure base directory for saving models and data
base_dir = 'model_outputs'
os.makedirs(base_dir, exist_ok=True)

# Iterate over each sector's index data
for sector, index in sector_indices.items():
    print(f"Processing sector: {sector}")
    
    # Create a sector-specific directory
    sector_dir = os.path.join(base_dir, sector)
    os.makedirs(sector_dir, exist_ok=True)
    
    # Convert to DataFrame for processing (if not already)
    data = index.copy()
    
    # Ensure 'Date' is the index if not already
    if 'Date' not in data.columns:
        data['Date'] = data.index
    data.set_index('Date', inplace=True)
    
    # Split the data into training (up to 2022) and testing (2023)
    train_data = data[data.index.year <= 2022]
    test_data = data[data.index >= '2022-10-01']  # Update to start from October 2022
    
    # Prepare features and target variable
    features = ['Returns', 'Volatility', 'MA_21', 'RSI', 'MACD']
    target = 'Future_Volatility'  # Forecasted volatility (target variable)
    
    # Check if the necessary columns exist
    if not all(f in train_data.columns for f in features):
        print(f"Skipping sector {sector} due to missing features.")
        continue
    
    # Drop rows with NaN values (in case any feature or target has NaNs)
    train_data.dropna(subset=features + [target], inplace=True)
    test_data.dropna(subset=features + [target], inplace=True)
    
    # Prepare training and testing data
    X_train = train_data[features]
    y_train = train_data[target]
    X_test = test_data[features]
    y_test = test_data[target]
    
    # --- Hyperparameter Tuning for Random Forest ---
    rf_model = RandomForestRegressor(random_state=42)
    rf_param_grid = {
        'n_estimators': [100, 200],
        'max_depth': [10, 20, None],
        'min_samples_split': [2, 5],
        'min_samples_leaf': [1, 2]
    }
    rf_grid_search = GridSearchCV(rf_model, rf_param_grid, cv=5, scoring='neg_mean_squared_error', n_jobs=-1)
    rf_grid_search.fit(X_train, y_train)
    rf_best_model = rf_grid_search.best_estimator_
    
    # --- Hyperparameter Tuning for MLP ---
    mlp_model = MLPRegressor(max_iter=1000, random_state=42)
    mlp_param_grid = {
        'hidden_layer_sizes': [(50, 50), (100,)],
        'activation': ['relu', 'tanh'],
        'solver': ['adam', 'sgd'],
        'alpha': [0.0001, 0.001],
    }
    mlp_grid_search = GridSearchCV(mlp_model, mlp_param_grid, cv=5, scoring='neg_mean_squared_error', n_jobs=-1)
    mlp_grid_search.fit(X_train, y_train)
    mlp_best_model = mlp_grid_search.best_estimator_
    
    # --- Predictions ---
    rf_pred = rf_best_model.predict(X_test)
    mlp_pred = mlp_best_model.predict(X_test)
    
    # --- Evaluate models ---
    rf_mse = mean_squared_error(y_test, rf_pred)
    rf_rmse = np.sqrt(rf_mse)
    mlp_mse = mean_squared_error(y_test, mlp_pred)
    mlp_rmse = np.sqrt(mlp_mse)
    
    print(f"RMSE for Random Forest model ({sector}): {rf_rmse:.4f}")
    print(f"RMSE for MLP model ({sector}): {mlp_rmse:.4f}")
    
    # Save the models
    rf_model_filename = os.path.join(sector_dir, f"{sector}_rf_model.pkl")
    mlp_model_filename = os.path.join(sector_dir, f"{sector}_mlp_model.pkl")
    
    # Save Random Forest model
    joblib.dump(rf_best_model, rf_model_filename)
    print(f"Random Forest model for {sector} saved as {rf_model_filename}")
    
    # Save MLP model
    joblib.dump(mlp_best_model, mlp_model_filename)
    print(f"MLP model for {sector} saved as {mlp_model_filename}")
    
    # Save test data (X_test and y_test) to CSV
    test_data_filename = os.path.join(sector_dir, f"{sector}_test_data.csv")
    
    # Combine X_test and y_test for saving
    test_data_combined = X_test.copy()
    test_data_combined['True_Volatility'] = y_test
    
    # Save to CSV
    test_data_combined.to_csv(test_data_filename, index=True)
    print(f"Test data for {sector} saved as {test_data_filename}")
    
    # Plot the predicted vs. actual volatility for both models in 2023
    plt.figure(figsize=(12, 6))
    plt.plot(test_data.index, y_test, label="Actual Volatility", color='blue', linestyle='-', linewidth=2)
    plt.plot(test_data.index, rf_pred, label="Random Forest Predicted Volatility", color='red', linestyle='--', linewidth=2)
    plt.plot(test_data.index, mlp_pred, label="MLP Predicted Volatility", color='green', linestyle='--', linewidth=2)
    
    plt.title(f'{sector} Actual vs Predicted Volatility (2023)', fontsize=14)
    plt.xlabel('Date', fontsize=12)
    plt.ylabel('Volatility', fontsize=12)
    plt.legend()
    plt.grid(True)
    plt.show()

# Now the models, test data, and predictions are saved in sector-specific folders
```

```python
# Define start and end dates for prediction
start_date = '2023-04-01'
end_date = '2023-06-30'
base_dir="model_outputs"
# Initialize dictionaries for storing predictions and actual values
predicted_volatility = {}
actual_volatility = {}

for sector, index in sector_indices.items():
    print(f"Predicting for sector: {sector}")
    
    # Create a sector-specific directory
    sector_dir = os.path.join(base_dir, sector)
    os.makedirs(sector_dir, exist_ok=True)
    
    # Convert to DataFrame for processing
    data = index.copy()
    
    # Ensure 'Date' is the index
    if 'Date' not in data.columns:
        data['Date'] = data.index
    data.set_index('Date', inplace=True)
    
    # Prepare features for prediction
    features = ['Returns', 'Volatility', 'MA_21', 'RSI', 'MACD']
    three_months_prior = pd.to_datetime(start_date) - pd.DateOffset(months=3)
    end_date_prior = pd.to_datetime(end_date) - pd.DateOffset(months=3)
    
    # Filter data for prediction using the defined start and end dates
    pred_input_data = data[(data.index >= three_months_prior) & (data.index <= end_date_prior)]
    
    if pred_input_data.empty or not all(f in pred_input_data.columns for f in features):
        print(f"Skipping sector {sector} due to insufficient prediction data.")
        continue
    
    X_pred = pred_input_data[features]
    X_pred.dropna(inplace=True)
    
    # Load the saved Random Forest model for the sector
    rf_model_filename = os.path.join(sector_dir, f"{sector}_rf_model.pkl")
    if not os.path.exists(rf_model_filename):
        print(f"No saved model found for {sector}, skipping...")
        continue
    
    rf_best_model = joblib.load(rf_model_filename)
    
    # Predict for the defined date range
    jan_mar_pred = rf_best_model.predict(X_pred)
    predicted_volatility[sector] = jan_mar_pred.tolist()
    
    # Get actual volatility for the defined date range
    jan_mar_actual_data = data[(data.index >= start_date) & (data.index <= end_date)]
    if not jan_mar_actual_data.empty:
        actual_volatility[sector] = jan_mar_actual_data['Future_Volatility'].dropna().tolist()
    else:
        actual_volatility[sector] = []

    print(f"Predictions for {sector} completed.")

# Save predictions and actual values as JSON
import json
results = {'predicted': predicted_volatility, 'actual': actual_volatility}

results_filename = os.path.join(base_dir, 'volatility_predictions.json')
with open(results_filename, 'w') as f:
    json.dump(results, f)

print(f"Predictions and actual volatilities saved to {results_filename}")
```

```python
import numpy as np
import pandas as pd
import scipy.optimize as sco
import matplotlib.pyplot as plt
import json
import joblib



# Assuming 'predicted_volatility' and 'actual_volatility' are dictionaries with sector volatilities (standard deviations)
# predicted_volatility and actual_volatility should be dictionaries in the same format

# Function to calculate portfolio volatility
def portfolio_volatility(weights, cov_matrix):
    return np.sqrt(np.dot(weights.T, np.dot(cov_matrix, weights)))

# Function to optimize portfolio based on minimum volatility
def minimize_volatility(expected_returns, cov_matrix):
    num_assets = len(expected_returns)
    initial_guess = np.ones(num_assets) / num_assets  # Initial guess: equal weights for all assets
    bounds = tuple((0, 1) for _ in range(num_assets))  # Bounds for weights (between 0 and 1)
    constraints = ({'type': 'eq', 'fun': lambda weights: np.sum(weights) - 1})  # Weights sum to 1

    # Minimize portfolio volatility (standard deviation)
    opt_result = sco.minimize(portfolio_volatility, initial_guess, args=(cov_matrix,), method='SLSQP', 
                              bounds=bounds, constraints=constraints)
    return opt_result

# Convert the volatility dictionaries into DataFrames for easier handling
predicted_volatility_df = pd.DataFrame(predicted_volatility)
actual_volatility_df = pd.DataFrame(actual_volatility)

# Assuming equal expected returns for simplicity (this can be replaced with actual expected return data)
expected_returns = np.ones(len(predicted_volatility_df.columns)) * 0.1  # 10% expected return for each sector

# Covariance matrices (simplified as diagonal for volatility, but you can compute actual covariance if available)
# Predicted Volatility Covariance Matrix
predicted_cov_matrix = np.diag(predicted_volatility_df.mean(axis=0) ** 2)

# Actual Volatility Covariance Matrix
actual_cov_matrix = np.diag(actual_volatility_df.mean(axis=0) ** 2)

# Perform portfolio optimization based on predicted volatility
optimization_result_predicted = minimize_volatility(expected_returns, predicted_cov_matrix)
optimized_weights_predicted = optimization_result_predicted.x
min_volatility_predicted = portfolio_volatility(optimized_weights_predicted, predicted_cov_matrix)

# Perform portfolio optimization based on actual volatility
optimization_result_actual = minimize_volatility(expected_returns, actual_cov_matrix)
optimized_weights_actual = optimization_result_actual.x
min_volatility_actual = portfolio_volatility(optimized_weights_actual, actual_cov_matrix)

# Assuming we want to compare both and use them for visualization
print("Optimized Weights based on Predicted Volatility:")
for sector, weight in zip(predicted_volatility_df.columns, optimized_weights_predicted):
    print(f"{sector}: {weight:.4f}")
print(f"Minimum Portfolio Volatility based on Predicted Volatility: {min_volatility_predicted:.4f}\n")

print("Optimized Weights based on Actual Volatility:")
for sector, weight in zip(actual_volatility_df.columns, optimized_weights_actual):
    print(f"{sector}: {weight:.4f}")
print(f"Minimum Portfolio Volatility based on Actual Volatility: {min_volatility_actual:.4f}\n")

# You can save the weights if needed (e.g., JSON file)
portfolio_weights = {
    'predicted_volatility': dict(zip(predicted_volatility_df.columns, optimized_weights_predicted)),
    'actual_volatility': dict(zip(actual_volatility_df.columns, optimized_weights_actual))
}

portfolio_filename = 'optimized_portfolio_weights.json'
with open(portfolio_filename, 'w') as f:
    json.dump(portfolio_weights, f)
print(f"Portfolio weights saved to {portfolio_filename}")

# Plotting the results to compare the optimized weights
labels = predicted_volatility_df.columns

# Plot for Predicted Volatility Portfolio
plt.figure(figsize=(10, 6))
plt.pie(optimized_weights_predicted, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title("Optimized Portfolio Weights Based on Predicted Volatility")
plt.show()
```

```python
def calculate_portfolio_return(start_date, end_date, portfolio_weights, sector_indices, total_portfolio_value):
    """
    Calculate the portfolio return over a specified date range.

    Parameters:
    - start_date (str): The start date in 'YYYY-MM-DD' format.
    - end_date (str): The end date in 'YYYY-MM-DD' format.
    - portfolio_weights (dict): A dictionary with sector names as keys and their weights as decimal values (e.g., 0.06 for 6%).
    - sector_indices (dict): A dictionary with sector names as keys and their corresponding DataFrames as values.
    - total_portfolio_value (float): The total value of the portfolio.

    Returns:
    - total_return (float): The total portfolio return over the specified period.
    """
    # Convert start_date and end_date to datetime
    start_date = pd.to_datetime(start_date)
    end_date = pd.to_datetime(end_date)

    # Step 1: Initialize total return
    total_return = 0.0

    # Step 2: Calculate investment amounts based on portfolio weights
    investments = {sector: total_portfolio_value * weight for sector, weight in portfolio_weights.items()}
    print("Investments in each sector:", investments)

    # Step 3: Iterate over each sector in the portfolio
    for sector, weight in portfolio_weights.items():
        if sector in sector_indices:
            # Step 4: Filter the sector index for the specified date range
            sector_data = sector_indices[sector]
            sector_data_filtered = sector_data[(sector_data.index >= start_date) & (sector_data.index <= end_date)]

            # Step 5: Calculate daily returns
            sector_data_filtered['Daily_Returns'] = sector_data_filtered['Close'].pct_change()

            # Step 6: Calculate weighted return for the sector
            weighted_return = (sector_data_filtered['Daily_Returns'] * weight).sum()

            # Step 7: Accumulate the total return
            total_return += weighted_return

    return total_return
```

```python
def calculate_max_return(start_date, end_date, sector_indices):
    """
    Calculate the maximum possible return over a specified date range for all sectors.

    Parameters:
    - start_date (str): The start date in 'YYYY-MM-DD' format.
    - end_date (str): The end date in 'YYYY-MM-DD' format.
    - sector_indices (dict): A dictionary with sector names as keys and their corresponding DataFrames as values.

    Returns:
    - max_return (float): The maximum return over the specified period.
    """
    # Convert start_date and end_date to datetime
    start_date = pd.to_datetime(start_date)
    end_date = pd.to_datetime(end_date)

    # Initialize max return
    max_return = float('-inf')  # Start with the lowest possible value

    # Iterate over each sector in the sector indices
    for sector, data in sector_indices.items():
        # Filter the sector index for the specified date range
        sector_data_filtered = data[(data.index >= start_date) & (data.index <= end_date)]

        # Calculate daily returns
        sector_data_filtered['Daily_Returns'] = sector_data_filtered['Close'].pct_change()

        # Calculate cumulative return over the period
        cumulative_return = (1 + sector_data_filtered['Daily_Returns']).prod() - 1  # Cumulative return formula

        # Update max return if the current sector's return is greater
        if cumulative_return > max_return:
            max_return = cumulative_return

    return max_return
```

```python
# Define month names for formatting
month_names = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']

# Extract month names from the start and end dates
start_month = month_names[int(start_date.split('-')[1]) - 1]  # Get month name from start_date
end_month = month_names[int(end_date.split('-')[1]) - 1]      # Get month name from end_date

# Calculate maximum return using the defined variables
max_return = calculate_max_return(start_date, end_date, sector_indices)

# Print the maximum return with formatted month names
print(f"Maximum possible return from {start_month} to {end_month} 2023: {max_return:.2%}")
```
